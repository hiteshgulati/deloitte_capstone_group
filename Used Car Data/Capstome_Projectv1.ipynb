```python
import pandas as pd
import numpy as np
```

```python
car_data = pd.read_csv('Car_sales_transactions.csv')
car_data.head()
```

```python
car_data.columns
```

```python
sales_status = pd.read_csv('Sales_Status.csv')
city_master = pd.read_excel('City_Master.xlsx')
car_data = pd.merge(car_data,sales_status, how = 'inner', on='Sales_ID')
car_data = pd.merge(car_data,city_master,how ='inner',on='City_Code')
car_data.head()
```

```python
car_data.info()
```

```python
car_data[['Mileage','Max_Power']]
```

```python
car_data['Mileage_float'] = car_data['Mileage'].str[:-5]
car_data['Power'] = car_data['Max_Power'].str[:-4]
```

```python
car_data['Mileage_float'] = car_data['Mileage_float'].apply(pd.to_numeric)
car_data['Power'] = car_data['Power'].apply(pd.to_numeric)
car_data['Power']
```

```python
car_data.info()
```

```python
car_data.groupby(['Name','Sold']).agg({'Mileage_float':np.mean,'Power':np.mean}).sort_values(['Name'],ascending = False)
```

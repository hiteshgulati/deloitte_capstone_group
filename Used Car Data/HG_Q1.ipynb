#### Import required Libraries

```python
import pandas as pd
import numpy as np
from IPython.display import display
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

#### Prepare the table

1) Import all csv / excel files
2) Merge files into single dataframe

```python
# 1) Import all csv.excel files
sales = pd.read_csv('Car_sales_transactions.csv')
city = pd.read_excel('City_Master.xlsx')
# postal_code = pd.read_excel('Postal_Code_Master.xlsx')
#TODO
region = pd.read_excel('Region_Master.xlsx')
region_state = pd.read_excel('Region_State_Mapping.xlsx')
sales_status = pd.read_csv('Sales_Status.csv')
state = pd.read_excel('State_Master.xlsx')

#2) Merge all files into single dataframe
df = sales.copy()
df = pd.merge(df,city[['City_Code','City_Name']],on='City_Code',how='left')
df = pd.merge(df,region_state,on='State_Code',how='left')
df = pd.merge(df,sales_status,on='Sales_ID',how='left')
df = pd.merge(df,state,on='State_Code',how='left')
df['Company'] = df['Name'].str.split(expand=True,pat=" ")[0]
df['Engine_cc'] = df['Engine'].str.split(pat=" ",expand=True)[0].astype(int)
df.head(3)
```

#### Queries

1) Which brands are selling most?
2) Are there specific locations selling more?
3) Which Fuel Types are selling most?
4) How Selling Price Varies by Engine Size


# 

```python
# 1) Which brands are selling most?

group_column = 'Company'
display_name = 'Brand'

group_df = df.groupby([group_column,'Sold'],as_index=False).agg({'Name':'count'})
group_df = group_df.pivot(index=group_column,columns='Sold',values='Name').fillna(0)
group_df['Total'] = group_df['N'] + group_df['Y']
group_df['% Sold'] = ((group_df['Y'] / group_df['Total'])*100).round(2)
group_df.sort_values(by='% Sold',ascending=False, inplace=True)
group_df.reset_index(inplace=True)
group_df.rename(columns={'Y':'Sold',group_column:display_name,'N':'Not Sold'},inplace=True)

display(group_df)
```

```python
# 2) Are there specific locations selling more?

group_column = 'City_Name'
display_name = 'City'

group_df = df.groupby([group_column,'Sold'],as_index=False).agg({'Name':'count'})
group_df = group_df.pivot(index=group_column,columns='Sold',values='Name').fillna(0)
group_df['Total'] = group_df['N'] + group_df['Y']
group_df['% Sold'] = ((group_df['Y'] / group_df['Total'])*100).round(2)
group_df.sort_values(by='% Sold',ascending=False, inplace=True)
group_df.reset_index(inplace=True)
group_df.rename(columns={'Y':'Sold',group_column:display_name,'N':'Not Sold'},inplace=True)

display(group_df)

```

```python
# 3) Which Fuel Types are selling more?

group_column = 'Fuel'
display_name = 'Fuel'

group_df = df.groupby([group_column,'Sold'],as_index=False).agg({'Name':'count'})
group_df = group_df.pivot(index=group_column,columns='Sold',values='Name').fillna(0)
group_df['Total'] = group_df['N'] + group_df['Y']
group_df['% Sold'] = ((group_df['Y'] / group_df['Total'])*100).round(2)
group_df.sort_values(by='% Sold',ascending=False, inplace=True)
group_df.reset_index(inplace=True)
group_df.rename(columns={'Y':'Sold',group_column:display_name,'N':'Not Sold'},inplace=True)

display(group_df)
```

```python
# 4) How Selling Price Varies by Engine Size

start = 600
end = 3800
bin_size = 200
bin_count = int((end-start)/bin_size)+1

bin_range = np.linspace(start,end,bin_count)

df['Engine_cc_bin'] = pd.cut(df['Engine_cc'], bins=bin_range).astype(str)

# df['Engine_cc_bin'] = [i.replace("(","").replace("]", "").split(", ") for i in df['Engine_cc_bin_int'].astype(str)]
# original_cuts = [pd.Interval(float(i), float(j)) for i, j in str_intervals]


group_column = 'Engine_cc_bin'
display_name = 'Engine'


slice_df = df[(df['Selling_Price'].notnull()) & (df[group_column].notnull())]
group_df = slice_df.groupby([group_column,'Sold'],as_index=False).agg({'Selling_Price':'mean'})
group_df = group_df.pivot(index=group_column,columns='Sold',values='Selling_Price').fillna(0)
# group_df['Total'] = group_df['N'] + group_df['Y']
# group_df['% Sold'] = ((group_df['Y'] / group_df['Total'])*100).round(2)
# group_df.sort_values(by='% Sold',ascending=False, inplace=True)
group_df.reset_index(inplace=True)
# group_df.rename(columns={'Y':'Sold in k',group_column:display_name,'N':'Not Sold in k'},inplace=True)
group_df['Y'] = (group_df['Y']/1000).astype(int)
group_df['N'] = (group_df['N']/1000).astype(int)
group_df.rename(columns={'Y':'Price (k) Sold',group_column:display_name,'N':'Price (k) Not Sold'},inplace=True)
group_df.sort_values(by='Engine',inplace=True)
display(group_df)
```

```python
iv = 'City_Name'
dv = 'Selling_Price'
model = ols(f'{dv} ~ C({iv})', data=df[df['Sold']=='Y'][[iv,dv]]).fit()
aov_table = sm.stats.anova_lm(model, typ=2) 
display(aov_table)

if aov_table.iloc[0]['PR(>F)'] < .025:
    print(f"We reject the hypothesis that {iv} don't have any impact on Selling Price")
else: 
    print(f"We cannot reject the hypothesis that {iv} don't have any impact on Selling Price")
```

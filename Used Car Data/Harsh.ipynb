## importing libraries

```python
import pandas as pd
import numpy as np
```

## importing all the required files

```python
sales = pd.read_csv('Car_sales_transactions.csv')
city = pd.read_excel('City_Master.xlsx')
region = pd.read_excel('Region_Master.xlsx')
region_state = pd.read_excel('Region_State_Mapping.xlsx')
sales_status = pd.read_csv('Sales_Status.csv')
state = pd.read_excel('State_Master.xlsx')
```

## merging the files into single datafram

```python
#2) Merge all files into single dataframe
df = sales.copy()
df = pd.merge(df,city[['City_Code','City_Name']],on='City_Code')
df = pd.merge(df,region_state,on='State_Code')
df = pd.merge(df,sales_status,on='Sales_ID')
df = pd.merge(df,state,on='State_Code')
df.head()
```

## adding new column with name 'Brand' in the dataframe

```python
df['Brand'] = df['Name'].str.split(' ').str[0]
df.head(5)
```

## filtering out the top 5 brands which is selling the most no. of cars

```python
df.query("Sold=='Y'").groupby('Brand', as_index=False).agg({'Sold':'count'}).rename(columns={'Sold':'Total_Sold'}).sort_values("Total_Sold", ascending=False).head()
```

## filtering out the top 10 cities which are selling most no. of cars

```python
df.query("City_Name=='Mysore'").groupby(['State_Name','City_Name'], as_index=False).agg({'Sold':'count'}).rename(columns={'Sold':'Total_Sold'}).sort_values("Total_Sold", ascending=False)
```

## finding average selling price based on brand

```python
# Average Selling Price based on brand
df.groupby('Brand', as_index=False).agg({'Sold':'count','Selling_Price':'mean'}).sort_values("Selling_Price", ascending=False).head()
```
